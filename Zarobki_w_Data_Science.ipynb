Stworzymy teraz 2 różne tabele. W jednej znajdą się pytania z dwiema cyframi, a w drugiej z jedną. Zabieg ten jest zrobiony, aby ułatwić wybór zmiennych (jeśli zmienna by się zaczynała od Q1, to zawierałaby także Q11, Q12, Q13,...). Następnie połączymy obie tabele.

```
#funkcja do stosowania startswith dla wielu kolumn
from pyspark.sql.functions import col, lit
from functools import reduce

def start():
  starts_with = reduce(
    lambda x, y: x | y,
    [col("value").startswith(s) for s in element_list], 
    lit(False))
  return starts_with
```

```
#zmiana dataframe na listę
columns_to_retain = columns_names.filter(start())
columns_to_retain = columns_to_retain.rdd.flatMap(lambda x: x).collect()
```

####Agregacja danych

Stwórzmy teraz ze zmiennej Salary zmienne pokazujące minimalną, maksymalną oraz średnią wartość zarobków danej osoby. Do porównań z innymi zmiennymi użyjemy wartości średniej, gdyż prawdopodobnie z najlepszym przybliżeniem pokazuje ona faktyczny stan zarobków ankietowanych.

Tworzenie nowych zmiennych z minimalnymi i maksymalnymi wartościami zarobków przebiegło pomyślnie. Kolumna Salary_min powinna być jednak przemnożona przez tysiąc, a kolumna Salary_max z kolei powinna pozbyć się znaku przecinka z wartości. Zmieńmy więc te wartości, a następnie zamieńmy jeszcze typy nowych kolumn na liczbowy. Następnie będziemy mogli stworzyć zmienną dotyczącą średniej wartości zarobków.

```
from pyspark.sql.functions import regexp_replace
data = data.withColumn('Salary_max', regexp_replace('Salary_max', ',', ''))
data = data.withColumn('Salary_min', data.Salary_min * 1000)
```

```
from pyspark.sql.types import IntegerType
data = data.withColumn("Salary_min", data.Salary_min.cast(IntegerType()))
data = data.withColumn("Salary_max", data.Salary_max.cast(IntegerType()))
data = data.withColumn('Salary_avg', (data.Salary_min + data.Salary_max) / 2)
```

```
display(data.select('Salary', 'Salary_min', 'Salary_max', 'Salary_avg').limit(10))
```

Możemy więc w końcu przejść do agregacji i wizualizacji danych.

Utwórzmy najpierw funkcje tworzące podsumowanie liczby unikatowych wartości dla danej kolumny oraz średnią wartość zarobków dla każdej z wartości danej cechy.

```
from pyspark.sql.functions import bround
from pyspark.sql.functions import desc

#funkcje dla zmiennych z pojedyncza kolumną
def val_counts(var):
  display(data.select(data.columns[var]).groupBy(data.columns[var]).count().sort(desc('count')))
def avg_salary(var):
  query = data.select(data.columns[var], 'Salary_avg').groupBy(data.columns[var]).avg('Salary_avg')
  display(query.select(data.columns[var], bround('avg(Salary_avg)').alias('Average salary')).sort((data.columns[var])))
  
#funkcje dla zmiennych znajdujących się w wielu kolumnach  
chart_names = ['Activities at work', 'Cloud providers', 'Programming languages', 
'Machine learning frameworks', 'Visualization tools', 'Relational databases', 
'Big Data and analytics', 'Data types']

def val_cols_counts(first_col, last_col):
  results = []
  for question in range(first_col, last_col):
    results.append([data.columns[question], data.select(data.columns[question]).filter(data[data.columns[question]] == data.columns[question]).count()])
    display(spark.createDataFrame(results).toDF(chart_names[chart], 'count').sort(desc('count')))    
  
def avg_cols_salary(first_col, last_col):
  results = []
  for question in range(first_col, last_col):
    query = data.select(data.columns[question], 'Salary_avg').filter(data[data.columns[question]] == data.columns[question]).groupBy(data.columns[question]).avg('Salary_avg')
    results.append([data.columns[question], query.select(bround('avg(Salary_avg)').alias('Average salary')).sort((data.columns[question])).collect()[0]['Average salary']])
    display(spark.createDataFrame(results).toDF(chart_names[chart], 'Average salary').sort(chart_names[chart]))    
```

### Relacyjne bazy danych

Niestety jak się okazuje, nazywanie kolumn według najczęściej występującej wartości poskutkowało utworzeniem się 2 kolumn SAP IQ. Aby pozbyć się tego problemu, stwórzmy z kopii danych 2 takie kolumny, ale z różnymi nazwami, a następnie dołączmy je do naszej głównej tabeli z danymi.

Kolumn o nazwie SAP IQ niestety nie możemy podmienić na 2 różne, gdyż są traktowane jako takie same i zmiany na jednej skutkują także takimi samymi zmianami na innej. Dodane nowe kolumny z kolei pojawią się na końcu zbioru danych. Najprościej więc nie korzystać w tym przypadku z funkcji i złączyć części zbioru ręcznie.

```
data_SAP = data_copy.select('Q29_Part_17', 'Q30_Part_15')
```

```
data = data.withColumn("id", monotonically_increasing_id())
data_SAP = data_SAP.withColumn("id", monotonically_increasing_id())

data.createOrReplaceTempView('data1')
data_SAP.createOrReplaceTempView('data2')

data = spark.sql("select * from data1 join data2 on data1.id = data2.id")
```

```
data = data.withColumnRenamed('Q29_Part_17', 'SAP IQ database')
data = data.withColumnRenamed('Q30_Part_15', 'SAP IQ analytics')

data = data.drop('id')
```

```
chart = chart + 1
results = []

for question in range(72, 88):
  results.append([data.columns[question], data.select(data.columns[question]).filter(data[data.columns[question]] == data.columns[question]).count()])

for question in range(89, 98):
  results.append([data.columns[question], data.select(data.columns[question]).filter(data[data.columns[question]] == data.columns[question]).count()])
  
results.append(['SAP IQ database', data.select(data['SAP IQ database']).filter(data['SAP IQ database'] == data['SAP IQ database']).count()])
  
display(spark.createDataFrame(results).toDF(chart_names[chart], 'count').sort(desc('count')))  
```

```
results = []

for question in range(72, 88):
  query = data.select(data.columns[question], 'Salary_avg').filter(data[data.columns[question]] == data.columns[question]).groupBy(data.columns[question]).avg('Salary_avg')
  results.append([data.columns[question], query.select(bround('avg(Salary_avg)').alias('Average salary')).sort((data.columns[question])).collect()[0]['Average salary']])
  
for question in range(89, 98):
  query = data.select(data.columns[question], 'Salary_avg').filter(data[data.columns[question]] == data.columns[question]).groupBy(data.columns[question]).avg('Salary_avg')
  results.append([data.columns[question], query.select(bround('avg(Salary_avg)').alias('Average salary')).sort((data.columns[question])).collect()[0]['Average salary']])
  
query = data.select(data['SAP IQ database'], 'Salary_avg').filter(data['SAP IQ database'] == data['SAP IQ database']).groupBy(data['SAP IQ database']).avg('Salary_avg')
results.append(['SAP IQ database', query.select(bround('avg(Salary_avg)').alias('Average salary')).sort((data['SAP IQ database'])).collect()[0]['Average salary']])  
  
display(spark.createDataFrame(results).toDF(chart_names[chart], 'Average salary').sort(desc('Average salary')))
```
